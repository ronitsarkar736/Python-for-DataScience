##### Q.8) Are battery electric vehicles (BEVs) typically more expensive than plug-in hybrid electric vehicles (PHEVs)?

```python
import scipy.stats as stats

# Assuming 'df' is your DataFrame and 'Electric Vehicle Type' and 'Base MSRP' are columns
# Filter data for BEVs and PHEVs
bevs = df[df['Electric Vehicle Type'] == 'Battery Electric Vehicle (BEV)']['Base MSRP']
phevs = df[df['Electric Vehicle Type'] == 'Plug-in Hybrid Electric Vehicle (PHEV)']['Base MSRP']

# Perform t-test
t_statistic, p_value = stats.ttest_ind(bevs, phevs, equal_var=False)

# Print t-test result
if p_value < 0.05:
    print("BEVs are typically more expensive than PHEVs (p < 0.05)")
else:
    print("There is no significant difference in prices between BEVs and PHEVs (p >= 0.05)")
```

##### Q.17) Is there a relationship between the type of electric vehicle and the county where it is registered?

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df['County'], df['Electric Vehicle Type'])

# Perform chi-square test of independence
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Display the test results
print(f"Chi-Square Statistic: {chi2}")
print(f"P-Value: {p}")

# Interpret the results
if p < 0.05:
    print("There is a significant relationship between the type of electric vehicle and the county where it is registered.")
else:
    print("There is no significant relationship between the type of electric vehicle and the county where it is registered.")
```

```python
# Identify categorical features
categorical_features = df.select_dtypes(include=['object']).columns.tolist()

# Check the number of unique categories in each categorical feature
unique_counts = df[categorical_features].nunique()

# Set a threshold for high cardinality
threshold = 100  # This can be adjusted based on specific needs

# Identify high-cardinality categorical features
high_cardinality_features = unique_counts[unique_counts > threshold].index.tolist()
print(f"High-Cardinality Features: {high_cardinality_features}")

# Frequency Encoding
for feature in high_cardinality_features:
    freq_encoding = df[feature].value_counts().to_dict()
    df[feature] = df[feature].map(freq_encoding)

# Grouping Rare Categories
# Define a function to group rare categories
def group_rare_categories(series, min_count):
    value_counts = series.value_counts()
    rare_categories = value_counts[value_counts < min_count].index
    series = series.apply(lambda x: 'Other' if x in rare_categories else x)
    return series

# Apply the function to each categorical feature
min_count = 100  # Minimum count to be considered a frequent category
for feature in categorical_features:
    df[feature] = group_rare_categories(df[feature], min_count)
```

##### Q.8) Dealing with Highly Correlated Variables: Are there pairs of variables that are highly correlated with each other? How will multicollinearity be addressed to prevent model instability?

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor
# Feature Selection and Correlation Analysis
# Compute the correlation matrix
corr_matrix = df.corr()

# Display the correlation matrix
print(corr_matrix)

# Plot the heatmap of the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()

# Identify highly correlated pairs (threshold > 0.75)
threshold = 0.75
high_corr_pairs = [(col1, col2) for col1 in corr_matrix.columns for col2 in corr_matrix.columns
                   if col1 != col2 and abs(corr_matrix.loc[col1, col2]) > threshold]

print("Highly Correlated Pairs (threshold > 0.75):", high_corr_pairs)

#  Multicollinearity Detection
# Calculate VIF for each feature
X = df.select_dtypes(include=[np.number])  # Select only numeric features
vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print("VIF for each feature:\n", vif_data)

# Addressing Multicollinearity
# Drop features with high VIF (example threshold > 10)
high_vif_features = vif_data[vif_data["VIF"] > 10]["Feature"].tolist()
df_reduced = df.drop(columns=high_vif_features)

print("Features with high VIF removed:", high_vif_features)
print(df_reduced.info())
```
